```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
file_path = 'Financial Analytics data.csv'  # Replace with your file path
df = pd.read_csv(file_path)
```

```python
df
```

```python
df.rename(columns={
    'Serial Number': 'Serial_Number',
    'Name': 'Company_Name',
    'Mar Cap - Crore': 'Market_Capitalization_Crore',
    'Sales Qtr - Crore': 'Quarterly_Sales_Crore'
    
},inplace=True)
```

```python
# Convert necessary columns to numeric if not already
df['Market_Capitalization_Crore'] = pd.to_numeric(df['Market_Capitalization_Crore'], errors='coerce')
df['Quarterly_Sales_Crore'] = pd.to_numeric(df['Quarterly_Sales_Crore'], errors='coerce')
```

```python
df
```

```python
# calculating sale- to market cap ratio
# Calculating Sales-to-Market Cap ratio
df['Sales_to_Market_Cap_Ratio'] = df['Quarterly_Sales_Crore'] / df['Market_Capitalization_Crore']
```

```python
# Correlation between Market Capitalization and Quarterly Sales
correlation = df['Market_Capitalization_Crore'].corr(df['Quarterly_Sales_Crore'])
print(f'Correlation between Market Capitalization and Quarterly Sales: {correlation}')
```

```python
# Plotting

# Scatter plot to visualize the relationship between Market Cap and Quarterly Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Market_Capitalization_Crore', y='Quarterly_Sales_Crore', data=df)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Market Capitalization (Crores)')
plt.ylabel('Quarterly Sales (Crores)')
plt.show()
```

```python
# Distribution of Sales-to-Market Cap Ratio
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales_to_Market_Cap_Ratio'], kde=True, bins=30)
plt.title('Distribution of Sales-to-Market Cap Ratio')
plt.xlabel('Sales-to-Market Cap Ratio')
plt.ylabel('Frequency')
plt.show()
```

```python
# Top 10 companies by Market Capitalization
top_10_market_cap = df.nlargest(10, 'Market_Capitalization_Crore')
print("Top 10 Companies by Market Capitalization:")
print(top_10_market_cap[['Company_Name', 'Market_Capitalization_Crore']])
```

```python
# Top 10 companies by Quarterly Sales
top_10_sales = df.nlargest(10, 'Market_Capitalization_Crore')
print("Top 10 Companies by Quarterly Sales:")
print(top_10_sales[['Company_Name', 'Quarterly_Sales_Crore']])
```

```python
# Bar plot for top 10 companies by Market Capitalization
plt.figure(figsize=(12, 6))
sns.barplot(x='Company_Name', y='Market_Capitalization_Crore', data=top_10_market_cap)
plt.title('Top 10 Companies by Market Capitalization')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for top 10 companies by Quarterly Sales
plt.figure(figsize=(12, 6))
sns.barplot(x='Company_Name', y='Quarterly_Sales_Crore', data=top_10_sales)
plt.title('Top 10 Companies by Quarterly Sales')
plt.xticks(rotation=90)
plt.show()
```
